```python
!pip install hmmlearn yfinance
import numpy as np
import datetime
import yfinance as yf 
from hmmlearn import hmm
import matplotlib.pyplot as plt 
from matplotlib.colors import ListedColormap 

plt.style.use("seaborn")
```

```python
# get vix data since beginning of 2020 to 9/12/2021 
today = datetime.date.today()
vix_data = yf.download("^VIX", start = "2020-01-01", end = today, interval = "1h")  

# separate dates for plotting and data for training 
vix_dates = vix_data.index  
vix_data = vix_data["Close"].values.reshape(-1, 1) 
```

```python
# assume there are 3 relevant hidden states
# volatility expansion, contraction, and lull 
model = hmm.GaussianHMM(n_components = 3, covariance_type = "full") 
model.fit(vix_data)  
```

```python
preds = model.predict(vix_data)
```

```python
plt.figure(figsize = (19, 13)) 

plt.subplot(211) 

colors = ListedColormap(['g',"r", "b"])  
classes = ["Volatility Lull", "Volatility Expansion", "Volatility Contraction"] 

scatter = plt.scatter(vix_dates, vix_data, c = preds, cmap = colors, alpha = 0.45)
plt.legend(handles = scatter.legend_elements()[0], labels = classes)
plt.plot(vix_dates, vix_data, "k")

plt.title("Hidden Markov Model VIX Volatility Regimes") 

plt.savefig("hmm_volatility_regimes.pdf", dpi = 150) 
plt.show() 
```
